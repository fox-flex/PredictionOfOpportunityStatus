```python
some = df[df['OpportunityId'] == '0061300001AAxCTAA1']
some = some[['OpportunityId', 'CreatedDateForInsert', 'PrevOpportunityStageUpdate', 'Amount', 'StageName', 'Probability']]
some['CreatedDateForInsert'] = pd.to_datetime(some['CreatedDateForInsert'],format='%m/%d/%Y %H:%M')
some['PrevOpportunityStageUpdate'] = pd.to_datetime(some['PrevOpportunityStageUpdate'],format='%m/%d/%Y %H:%M')
# tmp = tmp.groupby('date', as_index=False).agg({'electricity': 'sum'})

some = some.sort_values(by=['CreatedDateForInsert'])
some
```

```python
my = df[['OpportunityId', 'CreatedDateForInsert', 'StageName', 'Probability']]
my['CreatedDateForInsert'] = pd.to_datetime(my['CreatedDateForInsert'],format='%m/%d/%Y %H:%M')
my = my.sort_values(by=['CreatedDateForInsert'])
# my = my.groupby('OpportunityId')['Probability'].apply(list)

my = my.drop('CreatedDateForInsert', 1)
my = my.groupby('OpportunityId').agg(list)
my['StageName'] = my['StageName'].apply(lambda x : x[0])

# my = my.groupby('OpportunityId')[['StageName', 'Probability']].first()
# my = my.groupby('OpportunityId')[['StageName', 'Probability']]\
#         .apply(lambda x: x[['StageName','StageName']].values.tolist())\
#         .reset_index(name='col')
# my['Probability'] = my['Probability'][0]
```

```python
y = []
X = []
numX = 4

for lst in my['Probability']:
    if len(lst) < numX+1:
        continue
    for i in range(0, len(lst)-numX):
        y.append(lst[i+numX])
        X.append([lst[i+j] for j in range(numX)])
        

y = np.array(y)
X = np.array(X)

print(X.shape)
print(y.shape)
```
